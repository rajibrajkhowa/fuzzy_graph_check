```python
import random as rnd
import numpy as np
import itertools

# Create a list of vertices
VC = rnd.randint(3,10)
VL = list(range(1,VC+1,1))

# Create the vertex membership list
VM = []
A = np.random.random(VC)
for i in range(VC):
    VM.append(round(A[i],2))
VD = dict(zip(VL,VM))


# Create a list of edges
EL = list(itertools.combinations(VL,2))

# Create the edge membership list
EM = []
B = np.random.random(len(EL))/10
for i in range(len(EL)):
    EM.append(round(B[i],2))
ED = dict(zip(EL,EM))


# Check if the given vertices and membership values satisfies the fuzzy graph criteria

J =[]
for i in range(len(ED)):
    a = list(ED.values())[i]
    b = list(ED.keys())[i]
    c = b[0]
    d = b[1]
    e = VD.get(c)
    f = VD.get(d)
    g = min(e,f)
    if a<g:
        J.append(1)
    else:
        J.append(0)
  
if sum(J) == len(EL):
     print("The graph generated by Vertices: ", VL,"\nVertex membership: ", VD,"\nEdge membership: ", ED,"\nis a fuzzy graph")
else:
     print("The graph generated by Vertices: ", VL,"\nVertex membership: ", VD,"\nEdge membership: ", ED,"\nis not a fuzzy graph")

```
